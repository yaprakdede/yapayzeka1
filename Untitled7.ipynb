```python
import pandas as pd

#Ali nin ders ve not bilgileri
data = {'Ders' : ['Algoritma', 'BilgisayarProgramlama', 'Elektronik-2', 'AviyonikSistemler'],
        'Not': [95,100,65,90]}

#Veriyi bir Dataframe e dönüştür.
df= pd.DataFrame(data)

#en yüksek notu bul
en_yuksek_not = df['Not'].max()
print("En yüksek not:", en_yuksek_not)

#ortalama notu hesapla
ortalama_not = df['Not'].mean()
print("Ortalama not:", ortalama_not)

print(df.describe())
```

```python
import numpy as np

#Rastgale tanımlanan degerler - sıralaması olmayan
dizi = np.array([1, -55, 9, -23, 13, -21, 23, 37, 29, -6])

#Dizinin ort sonuç
ortalama = np.mean(dizi)
print("Ortalama:", ortalama)

#Dizinin standart sapması
standart_sapma = np.std(dizi)
print("Standart Sapma:", standart_sapma)

#Dizinin mutlağını al
mutlak_veriler = np.abs(dizi)
print("Mutlak deger cıktısı:", mutlak_veriler)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title("Sinüs Grafiği")
plt.xlabel("x ekseni")
plt.ylabel("y ekseni")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

#yıllara gore satışlar
yillar = [2018, 2019, 2020, 2021, 2022]
satislar = [150, 200, 20, -225, 14]

#satis grafigini çiz
plt.plot(yillar, satislar, marker='o')
#grafik baslıkları
plt.title("Yıllara Göre Satışlar")
plt.xlabel("yillar")
plt.ylabel("satis miktarı")
#grafigi goster
plt.show()
```
